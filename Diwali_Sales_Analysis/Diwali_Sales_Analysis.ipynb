```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount']= df['Amount'].astype('int')
```

Exploratory Data Analysis


```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df) 
```

```python
ax = sns.countplot(x = 'Gender', data = df) 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

From above Graphs we can see that most of the buyers are females and even the purchasing power of Female are greater that Men


```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

From above Bar Graph we can see that most of the buyers are of age group between 26-35yrs female


```python
df.columns
```

```python
sales_states= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data = sales_states)
```

```python
sales_states= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Amount', data = sales_states)
```

From above Graphs we can see that most of the orders & sales/amount are from Uttar Pradesh, Maharashtra, Karnataka

```python
ax =sns.countplot(data=df, x='Marital_Status')

sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_marital= df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data = sales_states, hue= 'Gender')
```

```python
ax =sns.countplot(data=df, x='Occupation')

sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_occ= df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data = sales_occ)
```

From above Graphs we can see that most of the buyers are working in IT, Healthcare and Aviation Sector

```python
ax =sns.countplot(data=df, x='Product_Category')

sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_product= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(x='Product_Category', y='Amount', data = sales_product)
```

#### From above Graphs we can see that most of the sold products are from Food, Clothing and Eloctronics



```python
sales_states= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID', y='Orders', data = sales_states)
```

### Married women age group 26-35yrs from UP, Maharshtra and Karnataka Working in IT, Healthcare and Aviation are more likely buy products from Food, Clothing and Electronics categories
